Its interesting to  compare what the first 200,000 data points to randomly selected sample data. It shows that it is important to  sample over picking a number of  data points. 

### Sampling Data

```python
np.random.seed(4)
```

```python
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
dev
```

```python
np.random.rand(10)
```

```python
big = ords_prods_merge[dev]
```

```python
small = ords_prods_merge[~dev]
```

```python
len(ords_prods_merge)
```

```python
len(big) + len(small)
```

```python
df_2 = small[['orders_day_of_week','prices']]
```

```python
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_prices_odow.png'))
```
